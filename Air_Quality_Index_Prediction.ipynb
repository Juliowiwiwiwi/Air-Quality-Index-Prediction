**Forecasting with FB Prophet Model**

```python
#Converting date fromat from dd/mm/yyy to yyy-mm-dd

date_info = pd.to_datetime(air_quality_data['Date'], dayfirst=True)
print(date_info)
```

```python
time_info=air_quality_data['Time']
print(time_info)
```

```python
time_info=time_info.apply(lambda x:x.replace('.',':'))
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

combining the 2 series(dateinfo and timeinfo) to a dataframe

```python
date_time=pd.concat([date_info,time_info],axis=1)
date_time.head()
```

```python
date_time.shape
```

combining date and time

```python
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
date_time.head()
```

```python
date_time.info()
```

Have to convert ds from object to datetime format

```python
data = pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

**Training the prophet model**

```python
model=Prophet()
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=365,freq='h')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=model.plot(forecast)
```

```python
fig2=model.plot_components(forecast)
```

```python
from prophet.diagnostics import cross_validation, performance_metrics
df_cv = cross_validation(model, initial='7200 h', period='720 h', horizon = '168 h')
df_p = performance_metrics(df_cv)
print("Cross-Validation Performance Metrics:")
print(df_p.head())
```

```python
date_to_predict = '2005-04-05 20:00:00'
future_data = pd.DataFrame({
    'ds': [date_to_predict]
})
future_data['ds'] = pd.to_datetime(future_data['ds'])
specific_forecast = model.predict(future_data)
print(f"\nPrediction for: {date_to_predict}")
print(f"Predicted Relative Humidity (yhat): {specific_forecast['yhat'].iloc[0]:.4f}")
print(f"Lower Bound (yhat_lower): {specific_forecast['yhat_lower'].iloc[0]:.4f}")
print(f"Upper Bound (yhat_upper): {specific_forecast['yhat_upper'].iloc[0]:.4f}")
```
